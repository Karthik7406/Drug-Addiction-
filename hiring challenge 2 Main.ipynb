```python

dataset = data.drop(columns=['Patient_ID','LocationDesc','Description','GeoLocation','Greater_Risk_Probability'])
```

```python
y = data['Greater_Risk_Probability']
y.head()
```

```python
dataset.head()
```


# Splitting the data into training and testing data


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(dataset,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```


# Model Selection

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(n_estimators=20,verbose=1)
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict = reg.predict(X_test)
```

# Evaluating the model

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
error = mean_squared_error(y_test,y_predict)
```

```python
error
```

```python
reg.score(X_train,y_train)
```

```python
r2_score(y_test,y_predict)
```

```python
print('Accuracy of the model',reg.score(X_train,y_train)*100,'%')
```
